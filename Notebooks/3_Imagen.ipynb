# Procesar imagen
Explorar librer√≠a en varias bandas

```python
import matplotlib.pyplot as plt
import cv2
import numpy as np
```

```python
imgpath = "Mosaicos/bosqueed.tif"
cb_img = cv2.imread(imgpath)
print("Dimensiones: ",cb_img.shape)
print("Tipo: ",cb_img.dtype)
```

```python
b,g,r= cv2.split(cb_img)
plt.figure(figsize=[20,5])
plt.subplot(141);plt.imshow(r,cmap='gray');plt.title("Rojo");
plt.subplot(142);plt.imshow(g,cmap='gray');plt.title("Verde");
plt.subplot(143);plt.imshow(b,cmap='gray');plt.title("Azul");
imgMerge = cv2.merge((b,g,r))
plt.subplot(144);plt.imshow(imgMerge[:,:,::-1]);plt.title("Merge");
```

```python
cb_img1 = cv2.cvtColor(cb_img, cv2.COLOR_BGR2HSV)
h,s,v= cv2.split(cb_img1)
plt.figure(figsize=[20,5])
plt.subplot(141);plt.imshow(h,cmap='gray');plt.title("H channel");
plt.subplot(142);plt.imshow(s,cmap='gray');plt.title("S channel");
plt.subplot(143);plt.imshow(v,cmap='gray');plt.title("V channel");
plt.subplot(144);plt.imshow(cb_img);plt.title("Original");
```

```python
h_new = h+100
img_merge2 = cv2.merge((h_new,s,v))
cb_img2 = cv2.cvtColor(img_merge2, cv2.COLOR_HSV2RGB)
plt.imshow(cb_img2);plt.title("Original");
```

```python
h_new = h + 255
s_new = s + 50
v_new = v + 50
img_merge2 = cv2.merge((h_new,s_new,v_new))
cb_img2 = cv2.cvtColor(img_merge2, cv2.COLOR_HSV2RGB)
plt.imshow(cb_img2);plt.title("Original");
```
